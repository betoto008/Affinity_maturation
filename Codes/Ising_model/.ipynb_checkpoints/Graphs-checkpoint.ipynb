```python
data=abs(np.loadtxt('output.txt'))
T=data[:,0]
E=data[:,1]
C=data[:,2]
M=data[:,3]
X=data[:,4]
```

```python
Tm=np.array([])
TM=np.array([])
index=0
for i in range(size(T)):
    if(T[i]<2.28):
        index+=1;
        Tm=np.append(Tm,-1*(T[i]-2.28))
    if(T[i]>2.28):
        TM=np.append(TM,(T[i]-2.28))
index, size(T)-index
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.loglog(TM,E[0:80],'g.', label='$T>T_{c}$' )
plt.loglog(Tm,E[80:150],'b.', label='$T<T_{c}$')
plt.legend()
plt.xlabel('|$T-T_{c}$|')
plt.ylabel('Energy per spin')
def func(x, a, b):
    return a*x**(b)
from scipy.optimize import curve_fit
popt, pcov = curve_fit(func, TM[-4:],E[-4:], bounds=(0, [10., 1.]))
a=popt[0]
b=popt[1]
plot(TM[-4:],func(TM[-4:],a,b))
TM[-4:]
plt.savefig('')
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.loglog(TM,C[0:80],'g.', label='$T>T_{c}$' )
plt.loglog(Tm,C[80:150],'b.', label='$T<T_{c}$')
plt.xlabel('|$T-T_{c}$|')
plt.ylabel('Heat Capacity per spin')
def funcC(x, a, b):
    return a*x**(-b)
from scipy.optimize import curve_fit
popt, pcov = curve_fit(funcC,Tm,C[80:150], bounds=(0, [20., 1.]))
popt2, pcov2 = curve_fit(funcC,TM,C[0:80], bounds=(0, [20., 1.]))
a=popt[0]
b=popt[1]
a2=popt2[0]
b2=popt2[1]
plt.loglog(Tm,funcC(Tm,a,b),label='$T<T_{c}$')
plt.loglog(TM,funcC(TM,a2,b2),label='$T>T_{c}$')
plt.legend()
eq='$C_{v} \sim |T-T_{c}|^{-0.215}$ for $T<T_{c}$ \n $C_{v} \sim |T-T_{c}|^{-0.247}$ for $T>T_{c}$'
plt.text(0.002,1,eq,fontsize=12)
plt.savefig('Alpha_fit.pdf')
b,b2
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.loglog(TM,M[0:80],'g.', label='$T>T_{c}$' )
plt.loglog(Tm,M[80:150],'b.', label='$T<T_{c}$')
plt.xlabel('|$T-T_{c}$|')
plt.ylabel('Magnetization per spin')
def func(x, a, b):
    return a*x**b
from scipy.optimize import curve_fit
popt, pcov = curve_fit(func,Tm,M[80:150], bounds=(-5., [20., 1.]))
a=popt[0]
b=popt[1]
eq='$M \sim |T-T_{c}|^{0.168}$ for $T<T_{c}$'
plt.text(0.002,0.01,eq,fontsize=12)
plt.loglog(Tm,func(Tm,a,b),label='fit $T<T_{c}$')
plt.legend()
plt.savefig('Beta_fit.pdf')


b
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.loglog(TM,X[0:80],'g.', label='$T>T_{c}$' )
plt.loglog(Tm,X[80:150],'b.', label='$T<T_{c}$')
plt.xlabel('|$T-T_{c}$|')
plt.ylabel('Susceptibility per spin')

def funcG(x, a, b):
    return a*x**(-b)
from scipy.optimize import curve_fit
popt, pcov = curve_fit(funcG,Tm[4:],X[84:150], bounds=(0., [.9, 2.]))
popt2, pcov2 = curve_fit(funcG,TM[:-4],X[0:76], bounds=(0., [.9, 2.]))
a=popt[0]
b=popt[1]
a2=popt2[0]
b2=popt2[1]
plt.loglog(Tm[4:],funcG(Tm[4:],a,b),label='$T<T_{c}$')
plt.loglog(TM[:-4],funcG(TM[:-4],a2,b2),label='$T>T_{c}$')
eq='$\chi \sim |T-T_{c}|^{-1.22}$ for $T<T_{c}$ \n $\chi \sim |T-T_{c}|^{-1.52}$ for $T>T_{c}$'
plt.text(0.002,4,eq,fontsize=12)

plt.legend()
plt.savefig('Gamma_fit.pdf')



b,b2
```
